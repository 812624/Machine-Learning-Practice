```python
# Importing all necessary packages
import pandas as pd
import numpy as np
from sklearn import  linear_model
from sklearn.cross_validation import train_test_split
import matplotlib.pyplot as plt
```

```python
# Opening .csv file using pandas
file = "/home/sakshi/Desktop/mlp/testdata.csv"
data = pd.read_csv(file)
data
```

```python
# Assigning single feature to data_x
data_x = data['X']

# Reshaping data if data has a single feature
data_x = data_x.values.reshape(-1,1)

# Assigning corresponding y value to data_y
data_y = data['Y']

# Applying linear regression
linReg = linear_model.LinearRegression()

# Dividing data in x_train, y_train and x_test, y_test in standard ratio of 0.8+0.2 
x_train, x_test, y_train, y_test = train_test_split(data_x, data_y, test_size=0.2)
linReg.fit(x_train, y_train)
```

```python
# Sampling x_test
a = linReg.predict(x_test)
a[5]
```

```python
# Corresponding values of y_test for selected x_test
y_test
```

```python
# Plotting the data
plt.plot()
plt.scatter(x_train, y_train, color="b", label="train data")
plt.scatter(x_test, y_test, color="r", label="test data")
plt.plot(x_train, linReg.predict(x_train), color="black")
plt.legend(loc=7)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# Mean square error
np.mean((linReg.predict(x_test) - y_test)**2)
```
